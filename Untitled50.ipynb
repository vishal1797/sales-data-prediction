```python
data1=pd.get_dummies(data[['Brand','Product Type']])
data1.head()
```

```python
data1.shape
```

```python
data.replace(to_replace ='2,598.51', value =2598.51)
```

```python
data["Sales Amount Usd"] = [float(str(i).replace(",", "")) for i in data["Sales Amount Usd"]]
```

```python
data["Quantity"] = [float(str(i).replace(",", "")) for i in data["Quantity"]]
```

```python
data2=data.join(data1)
data2.head()
```

```python
data2.shape
```

```python
x=data2.drop(['Quantity','Brand','Product Type'],axis=1)
```

```python
x.shape
```

```python
x.head()
```

```python
y=data2['Quantity']
```

```python
y.shape
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
np.random.seed(42)
```

```python
from sklearn.linear_model import LinearRegression
np.random.seed(42)


model=LinearRegression()
np.random.seed(42)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
np.random.seed(42)

model1=RandomForestRegressor()
np.random.seed(42)
model1.fit(x_train,y_train)
model1.score(x_test,y_test)
```
